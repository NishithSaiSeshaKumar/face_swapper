```python
import pycuda.driver as cuda
cuda.init()

# Check the number of available GPUs
num_gpus = cuda.Device.count()
print(f'Number of GPUs available: {num_gpus}')

if num_gpus > 0:
    for i in range(num_gpus):
        gpu = cuda.Device(i)
        print(f'GPU {i}: {gpu.name()}')
else:
    print('No CUDA-capable GPU detected.')
```

```python
import re

def remove_version_constraints(yaml_file, output_file):
    with open(yaml_file, 'r') as file:
        content = file.read()

    # Regular expression to match version constraints
    content_modified = re.sub(r'([a-zA-Z0-9_\-]+)==?\d+(\.\d+)*', r'\1', content)

    with open(output_file, 'w') as file:
        file.write(content_modified)

    print(f"Version constraints removed and saved to {output_file}")

# Example usage:
input_yaml = 'C:/Users/NSSK/Desktop/FACESWAP/Faceswap/work39_backup.yml'
output_yaml = 'C:/Users/NSSK/Desktop/FACESWAP/Faceswap/work39_modified.yml'  # Path to save the modified YAML file
remove_version_constraints(input_yaml, output_yaml)
```
